```python
movies = pd.read_csv('bollywood.csv', index_col = 'movie')['lead']
movies.head(3)
```

```python
subs = pd.read_csv('subs.csv').squeeze('columns')
subs.head(3)
```

```python
# we can write new value, change old values etc like edit can be done to series object
marks = pd.Series({'maths': 100, 'english': 50})
marks
```

```python
# we can change values by all types like fancy indexing, normnal indexing etc
marks[0] = 101
marks
```

```python
marks['english'] = 100
marks
```

```python
# if the specified index is no there then it will add that one
marks['science'] = 80
marks
```

```python
marks['maths':'science'] = 20
marks
```

```python
# if we use custom indexing it will select including that one also remember it
marks[0:2] = 30
marks
```

```python
# fancy indexing edit
marks[['maths', 'science']] = 10
marks
```

```python
marks[['maths', 'science']] = [20, 60]
marks
```

```python
#  core python functions and numpy functionality is also avilable here
marks
```
