```python
# it will be best if we use custom functyion on it
# finding the no.of movies starts with 'a' in each group
def start_a(group):
    return group['Series_Title'].str.startswith('A').sum()
genres.apply(start_a)
```

```python
movies['IMDB_Rating'].rank(ascending = False)
```

```python
# finding ranking of each movie according to imdb rating group wise
def rank(group):
    group['group_rank'] =  group['IMDB_Rating'].rank(ascending = False)
    return group
genres.apply(rank)
```

```python
# finding normalised rating based on group
def normal_rating(group):
    group['normal'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min()) / (group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
    return group
genres.apply(normal_rating)
```

```python
ipl = pd.read_csv('deliveries.csv')
ipl.head(2)
```

```python
ipl.columns
```

```python
# finding the top10 batsman in terms of runs
batsman = ipl.groupby('batsman')
batsman['batsman_runs'].sum().sort_values(ascending = False).head(10)
```

```python
# finding the batsman with max no.of sixes
six = ipl[ipl['batsman_runs'] == 6]
six.groupby('batsman').size().sort_values(ascending = False)
```

```python
# find the batsman with most no.of 4's and 6's in last 5 overi
last = ipl[((ipl['batsman_runs'] == 4) | (ipl['batsman_runs'] == 6)) & (ipl['over'] > 15)]
last.groupby('batsman').size().sort_values(ascending = False).head(1).index[0]
```

```python
# find kohlis record aginst all teams
temp_df = ipl[(ipl['batsman'] == 'V Kohli')]
temp_df
```

```python
temp_df.groupby('bowling_team')['batsman_runs'].sum().reset_index()
```

```python
# finding the highest score of each and every batsman
def highest_score(batsman):
    temp_df = ipl[ipl['batsman'] == batsman]
    return temp_df.groupby('match_id')['batsman_runs'].sum().max()

highest_score('CH Gayle')
```
