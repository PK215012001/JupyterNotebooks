```python
# creating a dataframe using 2d list
info = [
    ['Pavan', 'Vinukonda', 25, 'Kurukshetra'],
    ['Abhishek', 'Guntur', 24, 'Surat'],
    ['Pardhu', 'Sattenpalley', 25, 'Bhopal']
]
pd.DataFrame(info, columns = ['name', 'place', 'age', 'college'])
```
