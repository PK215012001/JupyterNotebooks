```python
import numpy as np
import pandas as pd
```

```python
# value_counts in dataframe tells the frequency count of rows, but it is mainly used for columns only
ipl = pd.read_csv('ipl-matches.csv')
ipl.head(2)
```

```python
# which player has won most potm in finals and qualifiers, isdigit() gives a true for digit
ipl[~ipl['MatchNumber'].str.isdigit()]['Player_of_Match'].value_counts()
```

```python
ipl.head(2)
```

```python
# toss decision plot
ipl['TossDecision'].value_counts().plot(kind = 'pie')
```

```python
# how many matches each team has played, the reason we added because index is same for both
(ipl['Team1'].value_counts() + ipl['Team2'].value_counts()).sort_values(ascending = False)
```

```python
students = pd.DataFrame(
    {
        'name':['nitish','ankit','rupesh',np.nan,'mrityunjay',np.nan,'rishabh',np.nan,'aditya',np.nan],
        'college':['bit','iit','vit',np.nan,np.nan,'vlsi','ssit',np.nan,np.nan,'git'],
        'branch':['eee','it','cse',np.nan,'me','ce','civ','cse','bio',np.nan],
        'cgpa':[6.66,8.25,6.41,np.nan,5.6,9.0,7.4,10,7.4,np.nan],
        'package':[4,5,6,np.nan,6,7,8,9,np.nan,np.nan]

    }
)

students
```

```python
# when it comes to null values, if we use sort on that column which contains null values, they will move downward by default
students.sort_values('name')
```

```python
# we can change this behaviour also like shown
students.sort_values('name', na_position = 'first')
```

```python
# we can even do descending 
students.sort_values('name', ascending = False)
```

```python
# rank gives us the rank of the original data based on what we have given and is applicable only for series
batsman_runs = pd.read_csv('batsman_runs_ipl.csv')
batsman_runs.head(5)
```

```python
# above contaisn the total runs of batsman in the whole ipl
# now we can ranl them based on runs if we want
batsman_runs['batsman_run'].rank(ascending = False)
```

```python
# the above result tells us the rank based on runs like ashishreddy's rank in 166.5 based on total runs
# we can even add this as column
batsman_runs['rank'] = batsman_runs['batsman_run'].rank(ascending = False)
batsman_runs.sort_values('rank')
```
