```python
# we can even do this for dataframe also as shown where we want to find the distance between two points where we can apply it along rows also
points_df = pd.DataFrame(
    {
        '1st point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
    }
)

points_df
```

```python
def distance(row):
    point_a = row['1st point']
    point_b = row['2nd point']
    return ((point_b[0] - point_a[0]) ** 2 + (point_b[1] - point_a[1]) ** 2) ** 0.5
    
points_df['distance'] = points_df.apply(distance, axis = 1)
points_df
```
