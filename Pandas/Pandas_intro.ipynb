```python
# Series object contains values, index, name, default inices are numbers, but we can have our own index also as shown below
subjects = ['maths', 'english', 'social']
marks = [50, 60, 100]
pd.Series(marks, index = subjects)
```

```python
# we can even name our series object
marks_series = pd.Series(marks, index = subjects, name = 'Subject marks')
marks_series
```

```python
# some important attributes of series object
#no.of itesm in a series
marks_series.size
```

```python
# datatype
marks_series.dtype
```

```python
# name of the object
marks_series.name
```

```python
# contains all unique values or not
marks_series.is_unique
```

```python
# to get the indexes of the object
marks_series.index
```

```python
# to get the values of the object
marks_series.values
```

```python
# we can directly create from dict also
marks = {'maths': 100, 'Social': 50}
pd.Series(marks)
```
