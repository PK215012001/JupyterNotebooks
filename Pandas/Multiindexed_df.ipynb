```python
# covid data
deaths = pd.read_csv('time_series_covid19_deaths_global.csv')
confirmed = pd.read_csv('time_series_covid19_confirmed_global.csv')
```

```python
deaths.head()
```

```python
confirmed.head()
```

```python
#if we observe above both tables are wide data and it is very difficult to work
# conveting above data with columns as country, date, conrimed, deaths
deaths = deaths.melt(id_vars= ['Province/State', 'Country/Region', 'Lat', 'Long'], var_name = 'date', value_name = 'number')
deaths
```

```python
confirmed = confirmed.melt(id_vars= ['Province/State', 'Country/Region', 'Lat', 'Long'], var_name = 'date', value_name = 'number')
confirmed
```

```python
# now joining two tables
covid = confirmed.merge(deaths, how = 'inner', on = ['Province/State', 'Country/Region', 'Lat', 'Long', 'date'])
```

```python
covid.rename(columns = {'number_x': 'confirmed_cases', 'number_y' : 'death_cases'}, inplace = True)
```

```python
covid = covid[['Country/Region', 'date', 'confirmed_cases', 'death_cases']]
covid
```
