```python
# making some custoom indexed data
students_dict = {
            'name': ['pavan', 'abhishek', 'ch mani', 'kishore'],
            'marks' : [20, 30, 40, 50],
            'place' : ['vnk', 'vrp', 'ngv', 'pdr']
}
students = pd.DataFrame(students_dict)
students
```

```python
# setting the name as index column
students.set_index('name', inplace = True)
students
```

```python
# now to access row using custom index we can use loc
students.loc['pavan']
```

```python
# we can even use iloc for above also, but the index will be default intgers
students.iloc[0]
```

```python
# same slicing
students.loc['pavan' : 'ch mani']
```

```python
# fancy indexing also works
students.loc[['pavan', 'ch mani']]
```

```python
# we can even access both rows and columns also by using this loc and iloc and for columns also iloc numbers will be the dafult index
students.iloc[0:2, 0]
```

```python
# now doing some data filtering
ipl = pd.read_csv('ipl-matches.csv')
ipl.head(2)
```

```python
# finiding the each season winners
mask = (ipl['MatchNumber'] == 'Final')
ipl[mask][['Season', 'WinningTeam']]
```

```python
# how many matches have super over finishes occurred
mask = (ipl['SuperOver'] == 'Y')
ipl[mask].shape[0]
```

```python
# how many matches csk won in kolkata
mask = (ipl['WinningTeam'] == 'Chennai Super Kings') & (ipl['City'] == 'Kolkata')
ipl[mask].shape[0]
```

```python
# toss winner is match winner in percentage
mask = (ipl['TossWinner'] == ipl['WinningTeam'])
(ipl[mask].shape[0] / ipl.shape[0]) * 100
```

```python
movies.head(2)
```

```python
# movies with rating >8 and votes > 10000
mask = (movies['imdb_rating'] > 8) & (movies['imdb_votes'] > 10000)
movies[mask]['title_x'].shape[0]
```

```python
# action movies with rating higher than 7.5
# the thing here is movies have multiple genres, so we first have to filter which movies have action genre in them
# for this we will split genres in each movie
mask = (movies['genres'].str.split('|').apply(lambda x: 'Action' in x)) & (movies['imdb_rating'] > 7.5)
movies[mask].shape[0]
```

```python
movies['genres'].str.contains('Action').head(2)
```

```python
# we can put new column easily as shown below
movies['Country'] = 'India'
movies
```

```python
# we can change the datatypes of columns if we want to save space
ipl.info()
```

```python
# for exampl above, id can be stored in int32, so
ipl['ID'] = ipl['ID'].astype('int32')
ipl.info()
```

```python
# above we saved nearly 5kb
# there is data type called category in pandas which is nothing but can be used for ctegorical data(data which repeats a lot , saves space)
# example, above city, season, teams these all are ctegory data because we know they are fixed and they will conly repeat
ipl[['City', 'Season', 'Team1', 'Team2', 'Venue', 'TossWinner', 'TossDecision', 'SuperOver', 'WinningTeam']] = ipl[['City', 'Season', 'Team1', 'Team2', 'Venue', 'TossWinner', 'TossDecision', 'SuperOver', 'WinningTeam']].astype('category')
ipl.info()
```
