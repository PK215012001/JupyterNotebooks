```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as ex
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# let's say we want to plot a scatter plot between total_bill and tip
# using pandas
tips.plot(kind = 'scatter', x = 'total_bill', y = 'tip')
```

```python
# using matplotlib
plt.scatter(tips['total_bill'], tips['tip'])
```

```python
# we can also do this using seaborn but the use is we can add multivariate analysis very easy and some styling is done automatically
# two types, figure level and axes level
#axes level
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip')
```

```python
# figure level(recommended)
sns.relplot(kind = 'scatter', data = tips, x = 'total_bill', y = 'tip')
```

```python
# if we observe the seaborn graph has x and ylables without even mentioning also
# we can do multi analysis by using hue(which sepearted the above data based on the categorie)
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'sex')
```

```python
sns.relplot(kind = 'scatter', data = tips, x = 'total_bill', y = 'tip', hue = 'sex')
```

```python
# we can even add two more style(difference shown in marker styles) and size(shown in size of the bubble) like this we can have nearly 5 variableanalysis
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'sex', style = 'smoker', size = 'size')
```

```python
sns.relplot(kind = 'scatter', x = 'total_bill', y = 'tip', hue = 'sex', style = 'smoker', size = 'size', data = tips)
```
