```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbs
```

```python
iris = pd.read_csv('iris.csv')
iris.head()
```

```python
iris['Species'].unique()
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'])
```

```python
# above contains all three species dat, if we want different color for each type, we first have to assign some numbering to each species
iris['Species'] = iris['Species'].replace({'Iris-setosa' : 0, 'Iris-versicolor' : 1, 'Iris-virginica' : 2})
iris['Species'].unique()
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c = iris['Species'])
```

```python
# if we want to know which color represents which use
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c = iris['Species'])
plt.colorbar()
```

```python
# we can even specify different color maps according to matplotlib docs
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c = iris['Species'], cmap = 'jet')
plt.colorbar()
```

```python
# we can even specify the opacity by alpha value
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c = iris['Species'], cmap = 'jet', alpha = 0.5)
plt.colorbar()
```

```python
# we can even fix our desired size
plt.figure(figsize = (10, 10))
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c = iris['Species'], cmap = 'jet', alpha = 0.9)
plt.colorbar()
```

```python
batters = pd.read_csv('batter.csv')
batters.head(5)
```

```python
# sampling 25 random from top 100 batsman
sample_df = batters.head(100).sample(25)
sample_df.head(5)
```

```python
plt.scatter(sample_df['avg'], sample_df['strike_rate'])
```

```python
# the thing is in above we don't know which dot is which batsman, we can assign lables also 
x = [1, 2, 3, 4]
y = [5, 6, 7, 8]
plt.scatter(x, y)
# to assign lable, we have to do below
plt.text(1, 5, 'point 1')
plt.text(2, 6, 'point 2')
plt.text(3, 7, 'point 3')
plt.text(4, 8, 'point 4', fontdict = {'size' : 12, 'color' : 'brown'})
```

```python
plt.scatter(sample_df['avg'], sample_df['strike_rate'])
```

```python
plt.scatter(sample_df['avg'], sample_df['strike_rate'])
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

```python
# we can even draw horizantal or vertical lines, like i want vertical line above 130 
plt.scatter(sample_df['avg'], sample_df['strike_rate'])
plt.axhline(130, color = 'red')
plt.axvline(30, color = 'red')
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 1, sharex = True, figsize = (7, 5))

ax[0].scatter(sample_df['avg'], sample_df['strike_rate'])
ax[0].set_title('Avg vs Strike Rate')
ax[0].set_ylabel('Strike Rate')

ax[1].scatter(sample_df['avg'], sample_df['runs'])
ax[1].set_title('Avg vs Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
```

```python
# we can even plot 3d plots, for that first we have to create a figure and then add one subplot inside it 
fig = plt.figure(figsize = (10, 7))
ax = fig.add_subplot(1, 1, 1,projection = '3d')
ax.scatter3D(sample_df['avg'], sample_df['strike_rate'], sample_df['runs'], color = 'red')
ax.set_title('Batsman analysis')
ax.set_xlabel('Average')
ax.set_ylabel('Strike Rate')
ax.set_zlabel('Runs')
```

```python
# when it comes to 3d line, it will just join points the speciality is the line goes in depth wise also
x = [0, 1, 2]
y = [0, 1, 2]
z = [0, 1, 2]

fig = plt.figure(figsize = (10, 7))
ax = fig.add_subplot(1, 1, 1, projection = '3d')
ax.scatter3D(x, y, z, s = [100, 100, 100])
ax.plot3D(x, y, z, color = 'red')
```

```python
# plotting 3d surface plots
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

X,  Y = np.meshgrid(x, y)
z = X**2 + Y**2

fig = plt.figure(figsize = (11, 7))
ax = fig.add_subplot(1, 1, 1, projection = '3d')
# colormap represents where values are higher and lower
# for 3d plots first we have to store it in seperate variable then we have to pass that into colorbar
p = ax.plot_surface(X, Y, z, cmap = 'viridis')
fig.colorbar(p)
```
