```python
# now plotting on real data
batsman = pd.read_csv('sharma-kohli.csv')
```

```python
batsman
```

```python
# plotting sharmas runs
plt.plot(batsman['index'], batsman['RG Sharma'])
```

```python
# we can even plot one after another
plt.plot(batsman['index'], batsman['RG Sharma'])
plt.plot(batsman['index'], batsman['V Kohli'])
```

```python
# mentioning x xis, yaxis, title
plt.plot(batsman['index'], batsman['RG Sharma'])
plt.plot(batsman['index'], batsman['V Kohli'])
plt.title('Rohit vs Kohli')
plt.xlabel('Season')
plt.ylabel('Runs')
```

```python
# we can even change the linestyle, linewidth, marking points, marking points width etc
plt.plot(batsman['index'], batsman['RG Sharma'], color = '#FF00D2', linestyle = 'solid', linewidth = 2, marker = 'D', markersize = 6, label = 'Rohit')
plt.plot(batsman['index'], batsman['V Kohli'], color = '#42AB11', linestyle = 'dotted', linewidth = 3, marker = 'o', markersize = 5, label = 'Kohli')
plt.title('Rohit vs Kohli')
plt.xlabel('Season')
plt.ylabel('Runs')
plt.legend()
plt.grid()
```

```python
# plotting multiple bar charts side by side(it is not direct, we have to do some manipulation)
df = pd.read_csv('batsman_season_record.csv')
```

```python
df
```

```python
# no we want to plot bar charts for each player for each season side by side
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width = 0.2)
plt.bar(np.arange(df.shape[0]), df['2016'], width = 0.2)
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width = 0.2)
# now we want to replace x axis with original names
plt.xticks(np.arange(df.shape[0]), df['batsman'])
plt.show()
```

```python
# sometimes our names are so long that it's very difficult to get
colors = ['red red red red red', 'blue blue blue blue', 'green green green green', 'yellow yellow yellow yellow']
freq = [20, 30, 40, 50]
plt.bar(colors, freq)
```

```python
# to aviod this we can use this
colors = ['red red red red red', 'blue blue blue blue', 'green green green green', 'yellow yellow yellow yellow']
freq = [20, 30, 40, 50]
plt.bar(colors, freq)
plt.xticks(rotation = 'vertical')
plt.show()
```

```python
# stacked bar chart gives total contribution and individual contribution
plt.bar(df['batsman'], df['2015'], label = '2015')
plt.bar(df['batsman'], df['2016'], bottom = df['2015'], label = '2016')
plt.bar(df['batsman'], df['2017'], bottom = (df['2016'] + df['2015']), label = '2017')
plt.legend()
```
