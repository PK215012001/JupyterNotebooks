```python
import numpy as np
```

```python
a = np.random.randint(1, 100, 15)
a
```

```python
# we can sort by directly using sort method 
np.sort(a)
```

```python
# for 2d array we have to specify which axis we want to sort
b = a.reshape(5, 3)
b
```

```python
# this sorts each row sepearetly
np.sort(b)
```

```python
np.sort(b, axis = 0)
```

```python
np.sort(b, axis = 1)
```

```python
# we can append eleemnts same like list
np.append(a, 100)
```

```python
# for 2d we have to specify which axis, either row or column
np.append(b, [[0, 1, 2]], axis = 0)
```

```python
np.append(b, [[0], [1], [2], [3], [4]], axis = 1)
```
