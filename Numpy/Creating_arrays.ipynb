We can create a numpy array just by passing any sequence into array function

```python
import numpy as np
```

```python
a = np.array([1, 2, 3, 4])
a
```

```python
b = np.array((1, 2, 3, 4))
b
```

```python
# Checking the type of array
type(a)
```

```python
# we can even create 2d and 3d arrays also
a = np.array([[1, 2, 3], [4, 5, 6]])
a
```

```python
#finidng the dimensions of the array
a.ndim
```

```python
b = np.array([[[1,2], [3, 4]], [[5, 6], [7, 8]]])
b
```

```python
b.ndim
```

```python
# By default this took int32 as the data type
a.dtype
```

```python
# we can even specify the type of datatype we want from the start itself
a = np.array([1, 2, 3, 4], dtype = 'int16')
a.dtype
```

```python
# same as ranage function we can even create an array using arange function
a = np.arange(1, 11)
a
```

```python
# we can even reshape the array we want
a = np.arange(1, 13)
a.reshape(3, 4)
```

```python
a.reshape(2, 3, 2)
```

```python
# np.ones
a = np.ones((3, 4))
a
```

```python
# np.zeros
a = np.zeros((2, 3))
a
```

```python
# np.random
a = np.random.random((3, 4))
a
```

```python
# np.linspace(this gives us an array of the elements we mentioned including first and last one with equal spaces
a = np.linspace(1, 10, 20)
a
```

```python
# np.identity(gives identity matrix of the given dimensions
a = np.identity(3)
a
```

```python
#a.shape 
a.shape
```

```python
# a.dtype
a.dtype
```

```python
# gives the size of each elemnet
a.itemsize
```

```python
# gives total no.of items in the array"
a.size
```

```python
a.dtype
```
