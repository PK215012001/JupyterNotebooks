# we can do array operations like scalar and vector operation

```python
import numpy as np
```

```python
# Scalar operations
a = np.arange(10, 24).reshape(2, 7)
a
```

```python
# arithemtic operations
a + 2
```

```python
a - 2
```

```python
a* 2
```

```python
a/2
```

```python
# Relational operations
a > 5
```

```python
a >= 20
```

```python
# Vector operations, if both have same shape, then also operations will happen one by one
b = np.arange(20, 34).reshape(2, 7)
b
```

```python
a+b
```

```python
a-b
```

```python
a > b
```
