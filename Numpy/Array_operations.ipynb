Numpy provides so many functions, some of the main functions as follows

```python
a = np.random.random((3, 3))
a = np.round(a * 100)
a
```

```python
#max
np.max(a)
```

```python
#min
np.min(a)
```

```python
#sum
np.sum(a)
```

```python
#prod
np.prod(a)
```

```python
# we can even get the above values from certain column or row or block or any other thing if we want as follows
# axis is an argument where 0 indicates highest dimension, and so axis 1 means we are travelling columns direction,operate rows wise and collapse rows
np.min(a, axis = 1)
```

```python
# if we do it for 3d matrix then we will get like this
a = np.round(np.random.random((3, 3, 3)) * 100)
a
```

```python
# now this has 3 dimensions and here axis = 0 operations by collapsing depth and thus gives us a single matrix of rows and columns,
#axis = 1, operates along rows so it collapses columns in each matrix and gives us seperate matrices which contains only rows
# axis = 2, operates along columns so it collapses only rows and gives us seperate matrices which contains the values of columns only
np.min(a, axis = 0)
```

```python
np.min(a, axis = 1)
```

```python
# axis =2,operate along columns meaning collapse rows
np.min(a, axis = 2) 
```
